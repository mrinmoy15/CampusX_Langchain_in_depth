```python
from langchain_openai import ChatOpenAI
from langchain.agents import create_agent
from langchain_core.tools import tool
import requests
from langchain_community.tools import DuckDuckGoSearchRun
import langchainhub as hub
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
```

```python
load_dotenv()
```

```python
# built in tool
search_tool = DuckDuckGoSearchRun()
```

```python
# custom tool
@tool
def get_weather_data(city: str) -> str:
  """
  This function fetches the current weather data for a given city
  """
  url = f'https://api.weatherstack.com/current?access_key=1508d4456946b54ec7f40c384e34f6da&query={city}'

  response = requests.get(url)

  return response.json()
```

```python
llm = ChatOpenAI(model="gpt-4o", temperature=0)
```

```python
agent = create_agent(
    model=llm,
    tools=[search_tool, get_weather_data],
)
```

```python
response = agent.invoke({"messages": [("user", "Find the capital of Madhya Pradesh, then find its current weather condition")]})
print(response["messages"][-1].content)
```
